```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Q1 - Sales Data Analysis")

roll_number = 102317144
np.random.seed(roll_number)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

sales_data = np.random.randint(1000, 5001, size=(12, 4))
df = pd.DataFrame(sales_data, index=months, columns=categories)

print("Original Data (First 5 rows):")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())

print("\nTotal Sales per Category:")
print(df.sum())
print("\nTotal Sales per Month:")
print(df.sum(axis=1))

print("\nAverage Sales Growth per Category:")
print(df.pct_change().mean() * 100)

df['Total_Sales'] = df.sum(axis=1)
df['Growth_Rate'] = df['Total_Sales'].pct_change() * 100

if roll_number % 2 == 0:
    print("\nApplying 10% discount to Electronics")
    df['Electronics'] = df['Electronics'] * 0.9
else:
    print("\nApplying 15% discount to Clothing")
    df['Clothing'] = df['Clothing'] * 0.85

plt.figure(figsize=(12, 6))
for category in categories:
    plt.plot(months, df[category], marker='o', label=category)
plt.title('Monthly Sales Trends')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("Q2 - Array Operations")

array = np.array([[1, -2, 3], [-4, 5, -6]])

print("Original array:")
print(array)

print("\nAbsolute values:")
print(np.abs(array))

print("\nPercentiles of flattened array (25th, 50th, 75th):")
print(np.percentile(array.flatten(), [25, 50, 75]))

print("\nPercentiles of each column (25th, 50th, 75th):")
print(np.percentile(array, [25, 50, 75], axis=0))

print("\nPercentiles of each row (25th, 50th, 75th):")
print(np.percentile(array, [25, 50, 75], axis=1))

print("\nFlattened array statistics:")
print(f"Mean: {np.mean(array):.2f}")
print(f"Median: {np.median(array):.2f}")
print(f"Standard Deviation: {np.std(array):.2f}")

print("\nColumn statistics:")
for i, col in enumerate(array.T):
    print(f"\nColumn {i+1}:")
    print(f"Mean: {np.mean(col):.2f}")
    print(f"Median: {np.median(col):.2f}")
    print(f"Standard Deviation: {np.std(col):.2f}")

print("\nRow statistics:")
for i, row in enumerate(array):
    print(f"\nRow {i+1}:")
    print(f"Mean: {np.mean(row):.2f}")
    print(f"Median: {np.median(row):.2f}")
    print(f"Standard Deviation: {np.std(row):.2f}")
```

```python
print("Q4 - List Element Swap")

def swap_elements(lst, pos1, pos2):
    if not isinstance(lst, list):
        raise TypeError("Input must be a list")
    if pos1 >= len(lst) or pos2 >= len(lst):
        raise IndexError("Position index out of range")
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst

my_list = [1, 2, 3, 4, 5]
print("Original list:", my_list)
swapped_list = swap_elements(my_list, 0, 4)
print("List after swapping first and last elements:", swapped_list)
```

```python
print("Q5 - Set Element Swap")

def swap_set_elements(input_set, val1, val2):
    if val1 not in input_set or val2 not in input_set:
        raise ValueError("Both values must exist in the set")

    temp_list = list(input_set)
    idx1 = temp_list.index(val1)
    idx2 = temp_list.index(val2)

    print("Set converted to list:", temp_list)

    temp_list[idx1], temp_list[idx2] = temp_list[idx2], temp_list[idx1]
    print("List after swapping:", temp_list)

    return set(temp_list)

my_set = {1, 2, 3, 4, 5}
print("Original set:", my_set)
result_set = swap_set_elements(my_set, 1, 5)
print("Final set:", result_set)
```
